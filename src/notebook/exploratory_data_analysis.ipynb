```python
import pandas as pd
from IPython.display import display
```

```python
class InstagramAnalytics:
    def __init__(self):
        self.users_df = None
        self.likes_df = None
        self.follows_df = None
        self.photos_df = None
        self.tags_df = None
        self.photo_tags_df = None
        self.comments_df = None
    
    def load_data(self, datasets):
        self.users_df = pd.read_csv(datasets["users"])
        self.likes_df = pd.read_csv(datasets["likes"])
        self.follows_df = pd.read_csv(datasets["follows"])
        self.photos_df = pd.read_csv(datasets["photos"])
        self.tags_df = pd.read_csv(datasets["tags"])
        self.photo_tags_df = pd.read_csv(datasets["photo_tags"])
        self.comments_df = pd.read_csv(datasets["comments"])
    
    def exploratory_analysis(self):
        print("Users' EDA")
        users = self.users_df.head()
        # Set index
        users.set_index("id", inplace=True)
        display(users)
        print(users.isnull().sum())
        print(users.shape)
        print(users.info())
        print("Likes EDF")
        display(self.likes_df.head())
        print(self.likes_df.isnull().sum())
        print(self.likes_df.shape)
        print(self.likes_df.info())
        print("Follows EDF")
        display(self.follows_df.head())
        print(self.follows_df.isnull().sum())
        print(self.follows_df.shape)
        print(self.follows_df.info())
        print("Photos EDF")
        self.photos_df.set_index("id", inplace=True)
        display(self.photos_df.head())
        print(self.photos_df.isnull().sum())
        print(self.photos_df.shape)
        print(self.photos_df.info())
        print("Tags EDF")
        self.tags_df.set_index("id", inplace=True)
        display(self.tags_df.head())
        print(self.tags_df.isnull().sum())
        print(self.tags_df.shape)
        print(self.tags_df.info())
        print("Photo tags EDF")
        display(self.photo_tags_df.head())
        print(self.photo_tags_df.isnull().sum())
        print(self.photo_tags_df.shape)
        print(self.photo_tags_df.info())
        print("Comments EDF")
        self.comments_df.set_index("id", inplace=True)
        display(self.comments_df.head())
        print(self.comments_df.isnull().sum())
        print(self.comments_df.shape)
        print(self.comments_df.info())
        
        
```

```python
if __name__ == "__main__":
    datasets = {
        "users": "../../archive/users.csv",
        "likes": "../../archive/likes.csv",
        "follows": "../../archive/follows.csv",
        "photos": "../../archive/photos.csv",
        "tags": "../../archive/tags.csv",
        "photo_tags": "../../archive/photo_tags.csv",
        "comments": "../../archive/comments.csv"
    }

    analytics = InstagramAnalytics()
    analytics.load_data(datasets)
    analytics.exploratory_analysis()
```
